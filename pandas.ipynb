```python
# Séries, única coluna
# Pode ser criado a partir de listras, array do numpy ou coluna de dataframe

s1 = pd.Series([2,5,3,34,54,23,1,16])
print(s1)
```

```python
# Serie a partir de um array do numpy

array1 = np.array([2,5,3,34,23,1,16])
s2 = pd.Series(array1)
print(s2)
```

```python
# Renomear

dados.rename(columns={'duration':'duração', 'purpose':'propósito'})
```

```python
# Porém a alterção não é persistitida

dados.head(1)
```

```python
# Para persistir

dados.rename(columns={'duration':'duração', 'porpose':'propósito'},inplace=True)
```

```python
# Excluir colunas

dados.drop('checking_status', axis=1, inplace=True)
print(dados)
```
